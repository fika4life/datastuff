```python
import pandas as pd
import chartify
```

```python
df = pd.read_csv('./data/life_expectancy_years.csv')

df.head()
```

```python
url = 'https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv'
regions = pd.read_csv(url, index_col=0, usecols= ['name','region'])

regions.head()
```

```python
df2 = pd.merge(left = df, right= regions, left_on ='country', right_on = 'name', how='left')
```

```python
eu= df2[df2['region'] == 'Europe']
```

```python
eu.head()
```

```python
eu = eu.dropna()

tidy_eu = eu.melt(id_vars = ['country','region'], var_name = 'year', value_name = 'life_expectancy')
```

```python
tidy_eu.head()
```

```python
tidy_eu.set_index('year')
```
